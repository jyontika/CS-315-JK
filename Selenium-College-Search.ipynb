### Jyontika's College Search

```python
import selenium 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python

def getCollegeInfo(query):
    """Given a search query, open Google, search the college name, 
    and record ranking, search popularity, and endowment."""
    
    geoDisabled = webdriver.FirefoxOptions() #disable the location popup
    geoDisabled.set_preference("geo.enabled", False)
    browser = webdriver.Firefox(options=geoDisabled)

    browser.get("https://google.com")  # Open Google

    # Move cursor to search bar
    search_bar = browser.find_element(By.NAME, "q")
    search_bar.clear()

    # Perform search
    search_bar.send_keys(query)
    search_bar.submit()

    # Wait for the search results to load
    time.sleep(1)

    try: 
        # Extract Average Cost Post-Aid
        info = browser.find_elements(By.CLASS_NAME, "wfRghe")
        #info = info.replace.text('\n', ' ')  # format better

        traits = []
        for element in info:
            traits.append(element.text)

        #formatting
        after_aid = traits[0].replace("Avg cost after aid", " ").strip()
        gradrate = traits[1].replace("Graduation rate", " ").strip()
        acceptance = traits[2].replace("Acceptance rate", " ").strip()

        #Extract Notable Alumni
        notable_alumni  = browser.find_elements(By.CLASS_NAME, "Wr0c6d")

        alum_list = [] 
        for element in notable_alumni:
            alum_list.append(element.text)
            
            if element == 3:
                break

        #Address
        address = browser.find_element(By.CLASS_NAME, "Z1hOCe").text
        address = address.replace("Address:", "").strip()


    except:
        print(f"Couldn't find a result for {query}")
        notable_alumni = ""
        alum_list = []
        traits = []

    # Close the browser
    browser.quit()

    return {
        'Cost' : after_aid,
        'GradRate': gradrate,
        'Acceptance': acceptance,
        'Alumni': alum_list,
        'Address': address
    }

```

```python

collegeList = [] #to store colleges
for school in ["Hamilton College", "Franklin and Marshall College",
             "Smith College", "Lafayette College"]:
    
    results = getCollegeInfo(school)

    print(school + ":")
    print("Address: ", results["Address"])
    print("Graduation Rate: ", results["GradRate"])
    print("Acceptance Rate: ", results["Acceptance"])
    print("Average Cost After Aid: ", results["Cost"])
    print("Notable Alumni: ", results["Alumni"])
    print("\n")
    
```
